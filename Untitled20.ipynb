```python
s3=(turno_noche["ds"] >= pd.to_datetime('2020-5-1')) & (turno_noche["ds"] < pd.to_datetime('2020-8-1')) 
```

```python
d3=turno_noche[s3].assign(affectedness=1)
```

```python
d4=turno_noche[s4].assign(affectedness=2)
```

```python
s5=(turno_noche["ds"] >= pd.to_datetime('2020-10-1')) & (turno_noche["ds"] < pd.to_datetime('2020-12-1')) 
d5=turno_noche[s5].assign(affectedness=1)
s6=(turno_noche["ds"] >= pd.to_datetime('2020-12-1')) & (turno_noche["ds"] < pd.to_datetime('2021-2-1')) 
d6=turno_noche[s6].assign(affectedness=2)
s7=(turno_noche["ds"] >= pd.to_datetime('2021-2-1')) & (turno_noche["ds"] < pd.to_datetime('2021-3-1')) 
d7=turno_noche[s7].assign(affectedness=1)
s8=(turno_noche["ds"] >= pd.to_datetime('2021-3-1'))
d8=turno_noche[s8].assign(affectedness=0)
```

```python
s1t=(turno_tarde["ds"] < pd.to_datetime('2020-3-1'))
d1t=turno_tarde[s1].assign(affectedness=0)
s2t=(turno_tarde["ds"] >= pd.to_datetime('2020-3-1')) & (turno_tarde["ds"] < pd.to_datetime('2020-5-1')) 
d2t=turno_tarde[s2].assign(affectedness=3)
s3t=(turno_tarde["ds"] >= pd.to_datetime('2020-5-1')) & (turno_tarde["ds"] < pd.to_datetime('2020-8-1')) 
d3t=turno_tarde[s3].assign(affectedness=1)
s4t=(turno_tarde["ds"] >= pd.to_datetime('2020-8-1')) & (turno_tarde["ds"] < pd.to_datetime('2020-10-1')) 
d4t=turno_tarde[s4].assign(affectedness=2)
s5t=(turno_tarde["ds"] >= pd.to_datetime('2020-10-1')) & (turno_tarde["ds"] < pd.to_datetime('2020-12-1')) 
d5t=turno_tarde[s5].assign(affectedness=1)
s6t=(turno_tarde["ds"] >= pd.to_datetime('2020-12-1')) & (turno_tarde["ds"] < pd.to_datetime('2021-2-1')) 
d6t=turno_tarde[s6].assign(affectedness=2)
s7t=(turno_tarde["ds"] >= pd.to_datetime('2021-2-1')) & (turno_tarde["ds"] < pd.to_datetime('2021-3-1')) 
d7t=turno_tarde[s7].assign(affectedness=1)
s8t=(turno_tarde["ds"] >= pd.to_datetime('2021-3-1'))
d8t=turno_tarde[s8].assign(affectedness=0)
```

```python
d9=pd.concat([d1,d2,d3,d4,d5,d6,d7,d8])
```
